```python
from diffusers import QwenImageEditPlusPipeline
MODEL_ID='FireRedTeam/FireRed-Image-Edit-1.0'
print(f'⏳ Loading {MODEL_ID}...')
pipe=QwenImageEditPlusPipeline.from_pretrained(MODEL_ID,torch_dtype=torch.bfloat16)
pipe.to('cuda'); pipe.set_progress_bar_config(disable=None)
print(f'✅ Loaded! VRAM: {vram()}')
```

```python
import gradio as gr
import glob,time
from PIL import Image
from datetime import datetime

DR='/content/drive/MyDrive'

def browse_drive(fp):
    if not fp: fp=DR
    if not os.path.isabs(fp): fp=os.path.join(DR,fp)
    if not os.path.exists(fp): return [],f'❌ Not found: {fp}'
    exts=('*.png','*.jpg','*.jpeg','*.webp','*.bmp')
    files=[]
    for e in exts: files.extend(glob.glob(os.path.join(fp,e))); files.extend(glob.glob(os.path.join(fp,e.upper())))
    files=sorted(set(files))[:100]
    subs=[d for d in sorted(os.listdir(fp)) if os.path.isdir(os.path.join(fp,d)) and not d.startswith('.')]
    info=f'📂 {fp}\n🖼️ {len(files)} gambar\n'
    if subs: info+=f'📁 {', '.join(subs[:20])}'
    return files,info

def load_from_drive(fp):
    if not fp: return None,'❌ Path!'
    if not os.path.isabs(fp): fp=os.path.join(DR,fp)
    if not os.path.exists(fp): return None,f'❌ Not found: {fp}'
    try:
        img=Image.open(fp).convert('RGB'); return img,f'✅ {os.path.basename(fp)} ({img.size[0]}x{img.size[1]})'
    except Exception as e: return None,f'❌ {e}'

def infer(images,prompt,seed,cfg,steps):
    with torch.inference_mode():
        return pipe(image=images,prompt=prompt,generator=torch.Generator(device='cuda').manual_seed(int(seed)),
            true_cfg_scale=float(cfg),negative_prompt=' ',num_inference_steps=int(steps),num_images_per_prompt=1).images[0]

def edit_single(inp,prompt,seed,cfg,steps,save):
    if inp is None: return None,'❌ Upload gambar!'
    if not prompt.strip(): return None,'❌ Tulis prompt!'
    t=time.time()
    try:
        out=infer([Image.fromarray(inp).convert('RGB')],prompt,seed,cfg,steps)
        s=f'✅ Done! ({time.time()-t:.1f}s)\n'
        if save:
            p=os.path.join(DOUT,f'firered_{datetime.now().strftime("%Y%m%d_%H%M%S")}_s{int(seed)}.png'); out.save(p); s+=f'💾 {p}\n'
        s+=f'VRAM: {vram()}'; return out,s
    except RuntimeError as e:
        if 'out of memory' in str(e).lower(): clear_mem(); return None,'❌ OOM!'
        return None,f'❌ {e}'

def edit_multi(i1,i2,i3,prompt,seed,cfg,steps,save):
    imgs=[Image.fromarray(x).convert('RGB') for x in [i1,i2,i3] if x is not None]
    if not imgs: return None,'❌ Upload!'
    if not prompt.strip(): return None,'❌ Prompt!'
    t=time.time()
    try:
        out=infer(imgs,prompt,seed,cfg,steps)
        s=f'✅ ({time.time()-t:.1f}s) {len(imgs)} imgs\n'
        if save:
            p=os.path.join(DOUT,f'firered_multi_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png'); out.save(p); s+=f'💾 {p}\n'
        return out,s
    except RuntimeError as e:
        if 'out of memory' in str(e).lower(): clear_mem(); return None,'❌ OOM!'
        return None,f'❌ {e}'

def bulk_edit(inf,outf,prompt,seed,cfg,steps,progress=gr.Progress()):
    if not prompt.strip(): return [],'❌ Prompt!'
    if not inf or not outf: return [],'❌ Folder!'
    inp=inf if os.path.isabs(inf) else os.path.join(DR,inf)
    outp=outf if os.path.isabs(outf) else os.path.join(DR,outf)
    if not os.path.exists(inp): return [],f'❌ {inp}'
    os.makedirs(outp,exist_ok=True)
    exts=('*.png','*.jpg','*.jpeg','*.webp','*.bmp')
    files=[]
    for e in exts: files.extend(glob.glob(os.path.join(inp,e))); files.extend(glob.glob(os.path.join(inp,e.upper())))
    files=sorted(set(files))
    if not files: return [],f'❌ No images'
    res=[]; log=f'📦 {len(files)} gambar | {prompt[:60]}\nIn: {inp}\nOut: {outp}\n\n'
    T=time.time(); ok=0; fail=0
    for i,fp in enumerate(progress.tqdm(files,desc='Bulk')):
        fn=os.path.basename(fp); log+=f'[{i+1}/{len(files)}] {fn}... '
        try:
            img=Image.open(fp).convert('RGB'); t=time.time()
            out=infer([img],prompt,seed,cfg,steps)
            out.save(os.path.join(outp,fn)); res.append(os.path.join(outp,fn)); ok+=1
            log+=f'✅ ({time.time()-t:.1f}s)\n'
        except: clear_mem(); log+='❌\n'; fail+=1
    el=time.time()-T
    log+=f'\n{"="*40}\n🎉 {ok}/{len(files)} | ⏱️ {el:.0f}s\n💾 {outp}'
    return res,log

def multi_prompt_single(inp,prompts_text,seed,cfg,steps,outf,progress=gr.Progress()):
    """1 gambar × banyak prompt."""
    if inp is None: return [],'❌ Upload gambar!'
    prompts=[p.strip() for p in prompts_text.strip().split('\n') if p.strip()]
    if not prompts: return [],'❌ Tulis minimal 1 prompt (1 per baris)!'
    outp=outf if os.path.isabs(outf) else os.path.join(DR,outf)
    os.makedirs(outp,exist_ok=True)
    img=Image.fromarray(inp).convert('RGB')
    res=[]; log=f'🎨 {len(prompts)} prompt × 1 gambar\nOut: {outp}\n\n'
    T=time.time(); ok=0
    for i,p in enumerate(progress.tqdm(prompts,desc='Prompts')):
        log+=f'[{i+1}/{len(prompts)}] "{p[:50]}"... '
        try:
            t=time.time(); out=infer([img],p,seed,cfg,steps)
            fn=f'prompt_{i+1}.png'; out.save(os.path.join(outp,fn))
            res.append(os.path.join(outp,fn)); ok+=1
            log+=f'✅ ({time.time()-t:.1f}s)\n'
        except: clear_mem(); log+='❌\n'
    log+=f'\n🎉 {ok}/{len(prompts)} | ⏱️ {time.time()-T:.0f}s\n💾 {outp}'
    return res,log

def multi_prompt_bulk(inf,prompts_text,seed,cfg,steps,outf,progress=gr.Progress()):
    """Banyak gambar × banyak prompt. Output: outf/prompt_N/filename.png"""
    if not inf: return [],'❌ Folder input!'
    prompts=[p.strip() for p in prompts_text.strip().split('\n') if p.strip()]
    if not prompts: return [],'❌ Tulis prompt!'
    inp=inf if os.path.isabs(inf) else os.path.join(DR,inf)
    outp=outf if os.path.isabs(outf) else os.path.join(DR,outf)
    if not os.path.exists(inp): return [],f'❌ {inp}'
    exts=('*.png','*.jpg','*.jpeg','*.webp','*.bmp')
    files=[]
    for e in exts: files.extend(glob.glob(os.path.join(inp,e))); files.extend(glob.glob(os.path.join(inp,e.upper())))
    files=sorted(set(files))
    if not files: return [],'❌ No images'
    total=len(files)*len(prompts)
    log=f'📦 {len(files)} gambar × {len(prompts)} prompt = {total} output\nIn: {inp}\nOut: {outp}\n\n'
    res=[]; T=time.time(); ok=0; n=0
    for pi,p in enumerate(prompts):
        pdir=os.path.join(outp,f'prompt_{pi+1}'); os.makedirs(pdir,exist_ok=True)
        log+=f'--- Prompt {pi+1}: "{p[:60]}" ---\n'
        for fi,fp in enumerate(progress.tqdm(files,desc=f'P{pi+1}')):
            fn=os.path.basename(fp); n+=1
            log+=f'  [{n}/{total}] {fn}... '
            try:
                img=Image.open(fp).convert('RGB'); t=time.time()
                out=infer([img],p,seed,cfg,steps)
                out.save(os.path.join(pdir,fn)); res.append(os.path.join(pdir,fn)); ok+=1
                log+=f'✅ ({time.time()-t:.1f}s)\n'
            except: clear_mem(); log+='❌\n'
    log+=f'\n{"="*40}\n🎉 {ok}/{total} | ⏱️ {time.time()-T:.0f}s\n💾 {outp}'
    return res,log

# ============ UI ============
THEME=gr.themes.Soft(primary_hue='red',secondary_hue='orange',neutral_hue='slate',font=[gr.themes.GoogleFont('Inter'),'sans-serif'])
CSS='.gradio-container{max-width:1200px!important}.mt{text-align:center}.mt h1{color:#dc2626;font-size:2em}.sb{font-family:monospace;font-size:.85em}'

with gr.Blocks(theme=THEME,css=CSS,title='FireRed Pro') as demo:
    gr.HTML("<div class='mt'><h1>🔥 FireRed Image Edit — Pro</h1><p style='color:#888'>bfloat16 · A100 · Drive I/O · Bulk · Multi-Prompt</p></div>")
    with gr.Tabs():
        with gr.TabItem('🖼️ Single Edit'):
            with gr.Row():
                with gr.Column():
                    si=gr.Image(label='Upload',type='numpy',height=400)
                    with gr.Accordion('📁 Load Drive',open=False):
                        sdp=gr.Textbox(label='Path',placeholder='Photos/foto.jpg')
                        slb=gr.Button('📂 Load',size='sm'); sls=gr.Textbox(interactive=False,lines=1)
                    sp=gr.Textbox(label='🎨 Prompt',lines=3,value='Transform the object into a miniature held between thumb and forefinger.')
                    with gr.Row(): ss=gr.Number(label='Seed',value=49,precision=0); sc=gr.Slider(label='CFG',minimum=1,maximum=10,value=4.0,step=0.5)
                    sst=gr.Slider(label='Steps',minimum=10,maximum=60,value=40,step=1)
                    ssv=gr.Checkbox(label='💾 Save',value=True); sb=gr.Button('🚀 Generate!',variant='primary',size='lg')
                with gr.Column():
                    so=gr.Image(label='Output',type='pil',height=400); sot=gr.Textbox(label='Status',interactive=False,lines=5,elem_classes='sb')
            slb.click(fn=load_from_drive,inputs=[sdp],outputs=[si,sls])
            sb.click(fn=edit_single,inputs=[si,sp,ss,sc,sst,ssv],outputs=[so,sot])

        with gr.TabItem('🖼️ Multi-Image'):
            gr.Markdown('### 2-3 gambar sekaligus')
            with gr.Row(): mi1=gr.Image(label='Img1',type='numpy',height=250); mi2=gr.Image(label='Img2',type='numpy',height=250); mi3=gr.Image(label='Img3',type='numpy',height=250)
            mp=gr.Textbox(label='🎨 Prompt',lines=3)
            with gr.Row(): ms=gr.Number(label='Seed',value=42,precision=0); mc=gr.Slider(label='CFG',minimum=1,maximum=10,value=4.0,step=0.5); mst2=gr.Slider(label='Steps',minimum=10,maximum=60,value=40,step=1)
            msv=gr.Checkbox(label='💾 Save',value=True); mbn=gr.Button('🚀 Generate!',variant='primary',size='lg')
            with gr.Row(): mo=gr.Image(label='Output',type='pil',height=400); mot=gr.Textbox(label='Status',interactive=False,lines=4)
            mbn.click(fn=edit_multi,inputs=[mi1,mi2,mi3,mp,ms,mc,mst2,msv],outputs=[mo,mot])

        with gr.TabItem('📦 Bulk Edit'):
            gr.Markdown('### 1 Prompt → Banyak Gambar\nOutput nama file **sama** seperti input.')
            with gr.Row(): bf=gr.Textbox(label='📂 Input Folder',placeholder='Photos/MyFolder',scale=2); bfp=gr.Button('🔍',scale=1)
            bfi=gr.Textbox(interactive=False,lines=2); bfg=gr.Gallery(columns=6,height=150,object_fit='contain')
            bof=gr.Textbox(label='📂 Output Folder',value='FireRed-Output/bulk')
            bp=gr.Textbox(label='🎨 Prompt',lines=2,value='Make it look like a watercolor painting.')
            with gr.Row(): bsd=gr.Number(label='Seed',value=42,precision=0); bcf=gr.Slider(label='CFG',minimum=1,maximum=10,value=4.0,step=0.5); bstp=gr.Slider(label='Steps',minimum=10,maximum=60,value=40,step=1)
            bb=gr.Button('📦 Start Bulk!',variant='primary',size='lg')
            bl=gr.Textbox(label='Log',interactive=False,lines=15,elem_classes='sb'); brs=gr.Gallery(columns=5,height=300,object_fit='contain')
            bfp.click(fn=browse_drive,inputs=[bf],outputs=[bfg,bfi])
            bb.click(fn=bulk_edit,inputs=[bf,bof,bp,bsd,bcf,bstp],outputs=[brs,bl])

        with gr.TabItem('🎨 Multi-Prompt'):
            gr.Markdown('### Banyak Prompt\n**Mode A:** 1 gambar × banyak prompt\n**Mode B:** Banyak gambar × banyak prompt\n\nTulis 1 prompt per baris. Output di subfolder `prompt_1/`, `prompt_2/`, dst.')
            with gr.Tabs():
                with gr.TabItem('1 Gambar × N Prompt'):
                    mps_img=gr.Image(label='Upload gambar',type='numpy',height=300)
                    mps_p=gr.Textbox(label='Prompts (1 per baris)',lines=6,placeholder='Make it watercolor\nMake it pencil sketch\nMake it pop art\nAdd snow everywhere')
                    with gr.Row(): mps_s=gr.Number(label='Seed',value=42,precision=0); mps_c=gr.Slider(label='CFG',minimum=1,maximum=10,value=4.0,step=0.5); mps_st=gr.Slider(label='Steps',minimum=10,maximum=60,value=40,step=1)
                    mps_o=gr.Textbox(label='📂 Output Folder',value='FireRed-Output/multi_prompt')
                    mps_b=gr.Button('🎨 Run Multi-Prompt!',variant='primary',size='lg')
                    mps_log=gr.Textbox(label='Log',interactive=False,lines=10,elem_classes='sb')
                    mps_gal=gr.Gallery(label='Results',columns=4,height=400,object_fit='contain')
                    mps_b.click(fn=multi_prompt_single,inputs=[mps_img,mps_p,mps_s,mps_c,mps_st,mps_o],outputs=[mps_gal,mps_log])

                with gr.TabItem('N Gambar × N Prompt'):
                    gr.Markdown('Setiap gambar di-edit dengan **setiap** prompt. Output: `output/prompt_1/img.png`, `output/prompt_2/img.png`, ...')
                    with gr.Row(): mpb_f=gr.Textbox(label='📂 Input Folder',placeholder='Photos/MyFolder',scale=2); mpb_prev=gr.Button('🔍',scale=1)
                    mpb_info=gr.Textbox(interactive=False,lines=2); mpb_gal_in=gr.Gallery(columns=6,height=150,object_fit='contain')
                    mpb_p=gr.Textbox(label='Prompts (1 per baris)',lines=6,placeholder='Make it watercolor\nMake it pencil sketch\nMake it pop art')
                    with gr.Row(): mpb_s=gr.Number(label='Seed',value=42,precision=0); mpb_c=gr.Slider(label='CFG',minimum=1,maximum=10,value=4.0,step=0.5); mpb_st=gr.Slider(label='Steps',minimum=10,maximum=60,value=40,step=1)
                    mpb_o=gr.Textbox(label='📂 Output Folder',value='FireRed-Output/multi_prompt_bulk')
                    mpb_b=gr.Button('📦🎨 Run Bulk Multi-Prompt!',variant='primary',size='lg')
                    mpb_log=gr.Textbox(label='Log',interactive=False,lines=15,elem_classes='sb')
                    mpb_gal=gr.Gallery(label='Results',columns=5,height=400,object_fit='contain')
                    mpb_prev.click(fn=browse_drive,inputs=[mpb_f],outputs=[mpb_gal_in,mpb_info])
                    mpb_b.click(fn=multi_prompt_bulk,inputs=[mpb_f,mpb_p,mpb_s,mpb_c,mpb_st,mpb_o],outputs=[mpb_gal,mpb_log])

        with gr.TabItem('📁 Drive'):
            with gr.Row(): gf=gr.Textbox(label='Folder',value='',scale=3); gb=gr.Button('🔍',variant='primary',scale=1)
            gi=gr.Textbox(interactive=False,lines=3); gg=gr.Gallery(columns=5,height=450,object_fit='contain')
            gb.click(fn=browse_drive,inputs=[gf],outputs=[gg,gi])

    gr.Markdown('---\n🔥 FireRed-Image-Edit-1.0')
demo.launch(share=True,debug=True)
```
